## Sklearn linear regression

```python
# import packages and modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# random dataset
df = pd.read_csv('starter_data/linear_regression_example_data.csv')
```

```python
#reshaping the data
X = df.iloc[:,0].values.reshape(-1,1)
Y = df.iloc[:,1].values.reshape(-1,1)

#create the linear regression environment
model = LinearRegression()

#fit the data that we want to use in the model
model.fit(X,Y)

# predict using the X values to get Y predictions
Y_predicted = model.predict(X)

#this is what our predicted values are
Y_predicted
```

```python

# visualize the regression model
plt.scatter(X,Y)
plt.plot(X, Y_predicted, color = 'red')
plt.show()
```
