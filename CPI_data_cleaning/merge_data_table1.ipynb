## Function that combines it together

```python
#function that combines the datasets
def combine_dataset(xlsx_lst):
    final_df = pd.DataFrame()
    
    #loops thru list of files and gets the CPI for categories
    for xlsx_month in xlsx_lst:
        
        #keep track of month_year in file name
        month_year = xlsx_month[-11:-5:1]
    
        #extra column index and header space
        min_df = pd.DataFrame(pd.read_excel('CPI_data/Table1/' + xlsx_month, index_col = [0], header = 3))
        min_df = min_df.dropna()
        
        #drop columns that are not needed
        if len(min_df.columns) < 10:
            min_df_column = min_df.columns[[0,1,2,4,5,6,7,8]]
            min_df = min_df.drop(columns = min_df_column, axis = 1)
            
            #set column name
            min_df.columns = [month_year]
            
        #some cases where there is a missing index column
        else:
            min_df_column = min_df.columns[[1,2,3,5,6,7,8,9]]
            min_df = min_df.drop(columns = min_df_column, axis = 1)
            
            #set index as the expenditure names
            min_df = min_df.set_index('Expenditure category')
            
            #set the columnn name as month and year
            min_df = min_df.rename(columns = {'Unnamed: 5': month_year})
        
        min_dfT = min_df.T
        
        #renaming the columns with parentheses
        #note to self: can improve the time complexity
        col = min_dfT.columns
        df_column_names = []
        for expenditure in col:
            #if there is parentheses, take those out. if not, keep name.
            if ('(' or ')') in expenditure:
                column_name = ''
                for letter in expenditure:
                    if (letter == '('):
                        break
                    else:
                        column_name = column_name + letter
                df_column_names.append(column_name)
            else:
                df_column_names.append(expenditure)
                
        #setting the renamed as columns
        min_dfT.columns = df_column_names
                
        #adding to final dataset
        final_df = final_df.append(min_dfT)
    return final_df
```

```python
#function that writes out all the file names with difference in years
def file_writer(year):
    #function that writes file names for a certain year
    lst = []
    for i in range(1,13): 
        
        #files before 2017 have a different file name
        if year < 2017:
            file_path = 'CpiPress1_FINAL_'
            if i < 10:
                
                # zero in front of months with one digit
                file_path = file_path + str(year) + '0' + str(i)
            else:
                file_path = file_path + str(year) + str(i)
            file_path = file_path + '.xlsx'
            lst.append(file_path)
        else:
            
            #file name for files after 2017
            file_path = 'news-release-table1-'
            if i < 10:
                file_path = file_path + str(year) + '0' + str(i)
            else:
                file_path = file_path + str(year) + str(i)
            file_path = file_path + '.xlsx'
            lst.append(file_path)
    return lst

def file_writer_years(start,end):
    #function that writes file names for all years
    lst_year = []
    complete_file_lst = []
    for i in range(start,end):
        lst_year.append(file_writer(i))
        
    #merge nested lists together
    for lst in lst_year:
        for val in lst:
            complete_file_lst.append(val)
    return complete_file_lst

#write file names for 2012 to 2022 (all the data we have)
filename_combine = file_writer_years(2012,2022)
```
