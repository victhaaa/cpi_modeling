```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

CPI_Merged = pd.read_csv('CPI_Data/Merged_Data/Merged_CPI.csv')
CPI_P = pd.read_csv('CPI_Data/Merged_Data/Monthly_CPI_data_w_percent_change.csv')
```

```python
# Delete first two rows of null values
CPI_Merged = CPI_Merged.iloc[2:,:]
```

```python
CPI_Merged
```

```python
# Merge two columns of airfares CPI into  one
Air_1 = CPI_Merged.iloc[:,-2][np.logical_not(np.isnan(CPI_Merged.iloc[:,-2]))]
Air_2 = CPI_Merged.iloc[:,-1][np.logical_not(np.isnan(CPI_Merged.iloc[:,-1]))]
Air_c = np.concatenate((Air_1, Air_2), axis = 0)
CPI_Merged = CPI_Merged.iloc[:,:-1]
CPI_Merged.iloc[:,-1] = Air_c
CPI_Merged.rename(columns={"Unnamed: 0":"Time"}, inplace = True)
CPI_Merged.set_index(['Time'], inplace = True)
CPI_Merged
```

```python
Cat_CPI = CPI_Merged.loc[:, ["Food","Energy","Apparel","New vehicles","Medical care commodities","Tobacco and smoking products","Shelter","Transportation services"]]
Cat_CPI.to_csv("CPI_Data/Cat_CPI&PPI/Cat_CPI.csv")
```

```python
fig, ax1 = plt.subplots(1, 1, figsize = (30,15))
Cat_CPI.plot(ax = ax1, fontsize = 13)
ax1.legend(loc = 2, fontsize = 15)
ax1.set_title("Categorized Consumer Price Index (CPI)", fontsize = 15)
ax1.set_xlabel("Time", fontsize = 15)
ax1.set_ylabel("CPI", fontsize = 15)




```

```python
Cat_PPI = pd.read_excel('CPI_Data/PPI/PPI2.xls', 'Monthly')
Cat_PPI.set_index('DATE', inplace = True)
Cat_PPI
```

```python
fig2, (ax2, ax3) = plt.subplots(1,2, figsize=(30,15))

Cat_CPI = Cat_CPI.loc[:, Cat_CPI.columns != "Tobacco and smoking products"]
Cat_CPI.plot(ax = ax2, fontsize = 13)
ax2.legend(loc = 2, fontsize = 15)
ax2.set_title("Categorized Consumer Price Index (CPI)", fontsize = 15)
ax2.set_xlabel("Time", fontsize = 15)
ax2.set_ylabel("CPI", fontsize = 15)

PPI_columns = ['Trade Service', 'Construction', 'Energy', 'Foods', 'Transportation and Warehouse Services', 'Final Demand Services Less Trade, Transportation, and Warehousing Services', 'Final Demand Goods Less Foods and Energy']
Cat_PPI.set_axis(PPI_columns, axis=1, inplace=True)
Cat_PPI.to_csv("CPI_Data/Cat_CPI&PPI/Cat_PPI.csv")
Cat_PPI.plot(ax = ax3)
ax3.legend(fontsize = 12)
ax3.set_title("Producer Price Index", fontsize = 15)
ax3.set_ylabel('PPI', fontsize = 13)
ax3.set_xlabel('Time', fontsize = 13)

```

```python
Cat_CPI.set_index(pd.to_datetime(Cat_CPI.index, format = "%Y%m"), inplace = True)
Cat_CPI.index
```

```python
Cat_PPI.index
```

```python
Merge_CPI_PPI = pd.merge(Cat_CPI, Cat_PPI, left_index=True, right_index=True)
Merge_CPI_PPI.to_csv("CPI_Data/Merged_Data/Merged_CPI_PPI.csv")
```

```python
Food_index = Merge_CPI_PPI[['Food', 'Foods']]
Food_index.columns = ['CPI', 'PPI']
Food_index.to_csv("CPI_Data/Cat_CPI&PPI/Food_Index.csv")
```

```python
Commodity_CPI = np.mean(Cat_CPI[['Apparel', 'New vehicles', 'Medical care commodities']], axis = 1)
Commodity_index = pd.concat((Commodity_CPI, Merge_CPI_PPI['Final Demand Goods Less Foods and Energy']), axis = 1)
Commodity_index.columns = ['CPI', 'PPI']
Commodity_index.to_csv("CPI_Data/Cat_CPI&PPI/Commodity_Index.csv")
```

```python
Energy_index = Merge_CPI_PPI[['Energy_x', 'Energy_y']]
Energy_index.columns = ['CPI', 'PPI']
Energy_index.to_csv("CPI_Data/Cat_CPI&PPI/Energy_Index.csv")
```

```python
Transportation_index = Merge_CPI_PPI[['Transportation services', 'Transportation and Warehouse Services']]
Transportation_index.columns = ['CPI', 'PPI']
Transportation_index.to_csv("CPI_Data/Cat_CPI&PPI/Transportation_Index.csv")
```
