```python
# make dataframe just the major expenditure categories
df = df[['Food and beverages','Housing','Apparel','Transportation',
        'Medical care','Recreation','Education and communication',
        'Other goods and services']]
```

```python
# create a function that checks correlation for all columns to get the best model columns
def evaluate_model(model, x, y):
    scores = []
    x = x.dropna()
    y = y.dropna()
    for x_column in x:
        y = y.drop(columns = [x_column], axis = 1)
        for y_column in y:
            x_train, x_test, y_train, y_test = train_test_split(x[[x_column]], y[y_column], random_state=1, test_size=0.2)
            model.fit(x_train, y_train)
            predict = model.predict(x_test)
            score = np.sqrt(metrics.mean_squared_error(predict, y_test))
            scores.append(score)
            index_name = x_column + ' ' + y_column
            print(f"The model {model} can predict {index_name} columns with root mean squared error of {score}")
```

```python
Linear_Reg = LinearRegression()
evaluate_model(Linear_Reg, df, df)
```

Food and Beverage column with Recreation has the lowest RMSE score out of all

```python
# run an accuracy score for the food and beverage column and recreation column
model = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(df[['Food and beverages']], df['Recreation'], random_state=1, test_size=0.2)
model.fit(x_train, y_train)
results = model.predict(x_test)
results
```

```python
r2_score = model.score(x_test,y_test)
r2_score
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(results, y_test)
```
