## Separate data in CPI: monthly cpi, annual percent change, monthly percent change

```python
#create a mini dataframe of all the annual percent change indexes
def annual_df(cpi):
    annual_lst = []
    
    #get all index for annual percent
    for i in cpi.index:
        if "Annual" in i:
            annual_lst.append(i)
    
    annual_mini_df = pd.DataFrame()
    #put index into dataframe
    for i in annual_lst:
        print(i)
        temp_df = pd.DataFrame(cpi[cpi.index == i])
        annual_mini_df = annual_mini_df.append(temp_df)
    return annual_mini_df

annual_df = annual_df(cpi)
annual_df
```

```python
# create a dataframe for the monthly changes
def monthly_df(cpi):
    monthly_lst = []
    
    #get all index for monthly percent
    for i in cpi.index:
        if "Monthly" in i:
            monthly_lst.append(i)
    
    monthly_mini_df = pd.DataFrame()
    #put index into dataframe
    for i in monthly_lst:
        print(i)
        temp_df = pd.DataFrame(cpi[cpi.index == i])
        monthly_mini_df = monthly_mini_df.append(temp_df)
    return monthly_mini_df

monthly_df = monthly_df(cpi)
monthly_df
```

The major expenditure categories that are used in CPI analysis is food and beverages, housing, apparel, transportation, medical care, recreation, education and communication, and other goods and services.
https://www.bls.gov/cpi/questions-and-answers.htm

```python
major_cpi_columns = ['Food and beverages','Housing','Apparel','Transportation',
                     'Medical care','Recreation','Education and communication','Other goods and services']
```

```python
# create a dataframe of just the major expenditure categories
major_cpi = cpi[major_cpi_columns]
major_cpi
```

```python
#plot the categories
major_cpi.plot(figsize = (20,10))
```

Looking at all the major categories...

- Food and beverages, Housing, Medical care, Recreation, Education, and Other goods and services follow a consistent linear trend

- Apparel has a consistent wavy trend which could be signifying consistent seasonality behavior

- Transportation does not have a consistent trend, it seems to be dipping and increasing at random intervals

```python
# looking into apparel trends
major_cpi['Apparel'].plot(style='.-', xticks = cpi.index, figsize = (20,10))
```

**There is always a consistent dip in the middle of the year for apparel and a then a consisten shoot back up to what it waas prior to the dip.**

Some potential reasons why this may be...
- prices are lowered during the summer months, then increase during times of gift giving?

There is also a prominent decrease between 2020 - 2021 which was influenced by the covid pandemic


```python
# looking into transporation trends
major_cpi['Transportation'].plot(style='.-', xticks = cpi.index, figsize = (20,10))
```

There are consistent dips usually at the beginning of the year.

- 2012, 2013, 2014, 2015, 2016, 2019, 2020 have these dips

Something to bring into question: Why would these decreases always be at the beginning of the year?
People are typically traveling for the holidays so the CPI prices should be higher? since flights near holidays are more expensive

Out of all the data, since the beginning of 2021 there has been a sky rocket in CPI for transportation
