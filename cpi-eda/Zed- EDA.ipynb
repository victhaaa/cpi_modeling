```python
df.columns[df.columns.str.contains('shelter') | df.columns.str.contains('Shelter')]
df_main = df.get(['Food', 'Energy', 'Apparel', 'Transportation', 'Medical care', 'Shelter', 'All items less food, shelter, energy, and used cars and trucks'])
df_main
```

```python
df_percent = df_main.loc[(df_main.index.str.contains('Annual Percent'))]
```

```python
df_percent
```

```python
df_percent = df_percent.assign(year = np.arange(2012,2022))
fig, (p1,p2) = plt.subplots(2, sharex=True, sharey=True)
p1.plot(df_percent.get('year'), df_percent.get(['Food', 'Energy', 'Apparel', 'Transportation']))
p2.plot(df_percent.get('year'), df_percent.get(['Medical care', 'Shelter', 'All items less food, shelter, energy, and used cars and trucks']))
p1.legend(['Food', 'Energy', 'Apparel', 'Transportation'])
p2.legend(['Medical care', 'Shelter', 'All items less food, shelter, energy, and used cars and trucks'])
p1.text(2014,-10,'boom in shale oil drilling')
p1.text(2020.2,15,'covid')
```

```python
df_food = df.loc[(df.index.str.contains('Annual Percent'))].get(df.columns[0:20]).assign(year = np.arange(2012,2022))
df_food
```

```python
fig, (p3,p4) = plt.subplots(2, sharex=True, sharey=True)

p3.plot(df_food.get('year'), df_food.get(df_food.columns[0:3]))
p3.legend(df_food.columns[0:3])
p4.plot(df_food.get('year'), df_food.get([df_food.columns[0], df_food.columns[14]]))
p4.legend([df_food.columns[0], df_food.columns[14]])
```
