# CPI sklearn model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Monthly_cpi_index.csv', index_col = 0)
df
```

```python
ppi = pd.read_csv('PPI_data.csv', index_col = 0).T
for i in ppi.isna().sum():
    print(i)
```

```python
#cant have null values in our model
df = df.dropna(axis = 1)
df
```

```python
#change index to datetime
df.index = pd.to_datetime(df.index)
```

```python
for i in df.columns:
    print(i)
```

```python
#create x and y variables
x = df.drop(columns = ['All items less food','All items less shelter','All items less food and shelter',
                        'All items less food, shelter, and energy','All items less food, shelter, energy, and used cars and trucks',
                       'All items less medical care','All items less energy'], axis = 1)
y = df['All items less food']
```

```python
#just the essential expenditure categories
x = df[['Food and beverages','Housing','Apparel','Transportation',
        'Medical care','Recreation','Education and communication',
        'Other goods and services']]

y = df['All items less food']
```

```python
#split data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# linear regression object
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predict = lr.predict(x_test)
```

```python
y_predict
```

```python
r2_score = lr.score(x_test,y_test)
r2_score
```
