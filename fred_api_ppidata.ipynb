```python
#atttempting to create dataframe that adds all API fetches
def fetch_data(series_dict):
    #input is a list of series id strings
    full_df = pd.DataFrame()
    
    for key,value in series_dict.items():
        #get the series dataframe
        mini_df = fred.get_series_df(key)
        
        #drop the columns of the retrieved dataframe
        mini_df = mini_df.drop(columns = ['realtime_start','realtime_end'], axis = 1)
        
        mini_df = mini_df.set_index('date')
        mini_df = mini_df.rename(columns = {'value':value})
        
        #wont append with indexes so used columns
        mini_dfT = mini_df.T
        
        full_df = full_df.append(mini_dfT)
    return full_df
```
